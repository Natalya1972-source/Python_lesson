```python
# 1. Квадрат числа
number = float(input("Введіть число: "))
square = number ** 2
print("Квадрат числа:", square)
```

```python
# 2. Середнє трьох чисел
a = float(input("Введіть перше число: "))
b = float(input("Введіть друге число: "))
c = float(input("Введіть третє число: "))

average = (a + b + c) / 3
print("Середнє:", average)
```

```python
# 3. Перетворення хвилин у години
minutes = int(input("Введіть кількість хвилин: "))

hours = minutes // 60
remaining_minutes = minutes % 60

print(hours, "години", remaining_minutes, "хвилин")
```

```python
# 4. Розрахунок знижки
price = float(input("Введіть ціну: "))
discount = float(input("Введіть знижку (%): "))

final_price = price * (1 - discount / 100)
print("Ціна зі знижкою:", final_price)
```

```python
# 5. Остання цифра числа
number = int(input("Введіть число: "))

last_digit = abs(number) % 10
print("Остання цифра:", last_digit)
```

```python
# 6. Периметр прямокутника
length = float(input("Введіть довжину: "))
width = float(input("Введіть ширину: "))

perimeter = 2 * (length + width)
print("Периметр:", perimeter)
```

```python
# 7. Виведення числа в стовпчик (4-значне число)

#  Рішення саме через // і %, як вимагають

number = int(input("Введіть 4-х значне число: "))

d1 = number // 1000
d2 = (number // 100) % 10
d3 = (number // 10) % 10
d4 = number % 10

print(d1)
print(d2)
print(d3)
print(d4)
```
