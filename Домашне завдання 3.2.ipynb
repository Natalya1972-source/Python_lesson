1 варіант 

```python
lst = [12, 3, 4, 10]   
if len(lst) == 0:
    result = lst
elif len(lst) == 1:
    result = lst
else:
    last_element = lst[-1]
    rest = lst[:-1]
    result = [last_element] + rest

print(result)
```
