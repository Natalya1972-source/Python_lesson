ДЗ 3.3. Розділити один список на два списки

```python
lst = [1, 2, 3, 4, 5, 6]   

if len(lst) == 0:
    result = [[], []]

elif len(lst) % 2 == 0:
    middle = len(lst) // 2
    first_part = lst[:middle]
    second_part = lst[middle:]
    result = [first_part, second_part]

else:
    middle = len(lst) // 2 + 1
    first_part = lst[:middle]
    second_part = lst[middle:]
    result = [first_part, second_part]

print(result)
```
