ДЗ 4.1. Перемістити всі нулі до кінця списку

1 варіант (цикл for) , відклала всі нулі, а потім додала їх в кінець списку

```python
lst = [0, 1, 0, 12, 3]

result = []
zeros = 0

for el in lst:
    if el == 0:
        zeros += 1
    else:
        result.append(el)

for i in range(zeros):
    result.append(0)

print(result)
```

Для всих одночасно 

```python
lsts = [
    [0, 1, 0, 12, 3],
    [0],
    [1, 0, 13, 0, 0, 0, 5],
    [9, 0, 7, 31, 0, 45, 0, 45, 0, 45, 0, 0, 96, 0]
]

for lst in lsts:
    result = []
    zeros = 0

    for el in lst:
        if el == 0:
            zeros += 1
        else:
            result.append(el)

    for _ in range(zeros):
        result.append(0)

    print(result)   
```

2 варіант (1 - забираю всі не нулі, 2- рахую скільки нулів, 3- створюю список з нулів, 4- склеюю списки разом)

```python
lst = [0, 1, 0, 12, 3]

result = [x for x in lst if x != 0] + [0] * lst.count(0)

print(result)
```

```python
lsts = [
    [0, 1, 0, 12, 3],
    [0],
    [1, 0, 13, 0, 0, 0, 5],
    [9, 0, 7, 31, 0, 45, 0, 45, 0, 45, 0, 0, 96, 0]
]

for lst in lsts:
    result = [x for x in lst if x != 0] + [0] * lst.count(0)
    print(result)
```

3. варіант (з циклом while )

```python
lsts = [
    [0, 1, 0, 12, 3],
    [0],
    [1, 0, 13, 0, 0, 0, 5],
    [9, 0, 7, 31, 0, 45, 0, 45, 0, 45, 0, 0, 96, 0]
]

for lst in lsts:
    index = 0
    length = len(lst)

    while index < length:
        if lst[index] == 0:
            lst.append(lst.pop(index))
            length -= 1
        else:
            index += 1

    print(lst)
```
