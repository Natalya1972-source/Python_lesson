ДЗ 4.2. Знайти суму елементів із парними індексами

1. через for

```python
def index_sum(lst):
    if not lst:
        return 0

    total = 0

    for index in range(len(lst)):
        if index % 2 == 0:
            total += lst[index]

    return total * lst[-1]
```

```python
print(index_sum([0, 1, 7, 2, 4, 8]))  
print(index_sum([1, 3, 5]))          
print(index_sum([6]))               
print(index_sum([]))                
```

```python
lst = [0, 1, 7, 2, 4, 8]   

total = 0

for i in range(0, len(lst), 2):
    total += lst[i]

if len(lst) == 0:
    result = 0
else:
    result = total * lst[-1]

print(result)
```

3. для всих списків одночасно

```python
lsts = [
    [0, 1, 7, 2, 4, 8],
    [1, 3, 5],
    [6],
    []
]

for lst in lsts:
    if len(lst) == 0:
        result = 0
    else:
        total = 0

        for i in range(0, len(lst), 2):
            total += lst[i]

        result = total * lst[-1]

    print(result)
```

4. кожен список окремо (цикл while)

```python
lst = [0, 1, 7, 2, 4, 8]

total = 0
i = 0

while i < len(lst):
    if i % 2 == 0:
        total += lst[i]
    i += 1

result = total * lst[-1]
print(result)   
```
