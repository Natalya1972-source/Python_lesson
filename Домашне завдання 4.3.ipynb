```python
ДЗ 4.3. Список із 3 елементів
```

1. варіант (через індексацію)

```python
import random

length = random.randint(3, 10)
numbers = [random.randint(1, 10) for _ in range(length)]

result = [numbers[0], numbers[2], numbers[-2]]

print(numbers, "==", result)
```

2. варіант (через append)

```python
import random

numbers = []

for _ in range(random.randint(3, 10)):
    numbers.append(random.randint(1, 10))

result = []
result.append(numbers[0])
result.append(numbers[2])
result.append(numbers[-2])

print(numbers, "==", result)
```

3. варіант (через зрізи списку)

```python
import random

numbers = [random.randint(1, 10) for _ in range(random.randint(3, 10))]

result = numbers[:3:2] + numbers[-2:-1]

print(numbers, "==", result)
```

4. варіант (через змінні)

```python
import random

numbers = [random.randint(1, 10) for _ in range(random.randint(3, 10))]

first = numbers[0]
third = numbers[2]
second_from_end = numbers[-2]

result = [first, third, second_from_end]

print(numbers, "==", result)
```

5. варіант (функція)

```python
import random

lst = [random.randint(1, 10) for _ in range(random.randint(3, 10))]
print(lst, "==", [lst[0], lst[2], lst[-2]])
```
